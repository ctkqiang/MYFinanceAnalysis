# 马来西亚单位信托基金

单位信托（Unit Trust）是一种投资基金的形式，由许多投资者共同投资于一个基金中。这个基金由专业的基金经理管理，他们将投资者的钱集中起来，投资于各种资产，如股票、债券或房地产。投资者会根据他们投资的份额（即单位）获得基金的收益。单位信托的优点包括分散投资风险和由专业人士进行投资管理，适合那些希望将资金交给专业人士管理的投资者。

```python
import pandas as pd

df = pd.read_csv("数据/fsm-fundscsv.csv")

column_mapping = {
    "Fund Name": "基金名称",
    "Date Launch": "上市日期",
    "Min. Initial Inv": "最低初始投资",
    "Market Sales Charges": "市场销售费用",
}

df.rename(columns=column_mapping, inplace=True)
display(df)
```

####  列出销售费用为0%的基金


```python
# 过滤销售费用为0%的基金
zero_sales_charge_funds = df[df["市场销售费用"] == "0%"]

# 显示结果
display(zero_sales_charge_funds)
```

#### 列出所有最低初始投资金额 <= 100 的基金

```python
# 去除货币符号和逗号，然后转换为数值
df["最低初始投资"] = (
    df["最低初始投资"]
    .replace({"MYR": "", "USD": "", "AUD": "", "SGD": "", ",": ""}, regex=True)
    .astype(float)
)

# 过滤投资金额 <= 100 的基金
min_initial_inv_100 = df[df["最低初始投资"] <= 100]

# 显示结果
display(min_initial_inv_100)
```
